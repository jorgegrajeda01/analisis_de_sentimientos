```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

```python
df = pd.read_csv("./data/IMDB Dataset.csv")
```

```python
df
```

```python
df.sample(5)
```

```python
df_pos = df[df['sentiment']=='positive'][:5000]
df_neg = df[df['sentiment']=='negative'][:5000]

df_reviews = pd.concat([df_pos, df_neg ])
```

```python
from sklearn.model_selection import train_test_split
```

```python
train,test = train_test_split(df_reviews,test_size =0.33,random_state=42)
```

```python
train_x, train_y = train['review'], train['sentiment']
test_x, test_y = test['review'], test['sentiment']
```

```python
train_y.value_counts()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(stop_words='english')
train_x_vector = tfidf.fit_transform(train_x)
test_x_vector = tfidf.transform(test_x)
```

```python
primera_resenia = pd.DataFrame.sparse.from_spmatrix(train_x_vector,
                                  index=train_x.index,
                                  columns=tfidf.get_feature_names_out()).iloc[0]
```

```python
from sklearn.svm import SVC
svc = SVC(kernel='linear')
svc.fit(train_x_vector, train_y)
```

```python
print(svc.predict(tfidf.transform(['A good movie'])))
print(svc.predict(tfidf.transform(['An excellent movie'])))
print(svc.predict(tfidf.transform(['I did not like this movie at all I gave this movie away'])))
```

```python
print(svc.score(test_x_vector, test_y))
```

```python
from sklearn.metrics import f1_score

f1_score(test_y,svc.predict(test_x_vector),
          labels = ['positive','negative'],average=None)
```

```python
from sklearn.metrics import classification_report

print(classification_report(test_y,
                            svc.predict(test_x_vector),
                            labels = ['positive','negative']))
```

```python
from sklearn.metrics import confusion_matrix

conf_mat = confusion_matrix(test_y,
                           svc.predict(test_x_vector),
                           labels = ['positive', 'negative'])
conf_mat
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
```

```python
lr = LogisticRegression()
lr.fit(train_x_vector, train_y)
```

```python
print("LogisticRegression Score:", lr.score(test_x_vector, test_y))
print("LogisticRegression F1:", f1_score(test_y, lr.predict(test_x_vector), labels=['positive','negative'], average=None))
print(classification_report(test_y, lr.predict(test_x_vector), labels=['positive','negative']))
print("Confusion Matrix LR:")
print(confusion_matrix(test_y, lr.predict(test_x_vector), labels=['positive', 'negative']))
```

```python
gnb = GaussianNB()
gnb.fit(train_x_vector.toarray(), train_y)
```

```python
print("GaussianNB Score:", gnb.score(test_x_vector.toarray(), test_y))
print("GaussianNB F1:", f1_score(test_y, gnb.predict(test_x_vector.toarray()), labels=['positive','negative'], average=None))
print(classification_report(test_y, gnb.predict(test_x_vector.toarray()), labels=['positive','negative']))
print("Confusion Matrix GNB:")
print(confusion_matrix(test_y, gnb.predict(test_x_vector.toarray()), labels=['positive', 'negative']))
```

```python
dt = DecisionTreeClassifier()
dt.fit(train_x_vector, train_y)
```

```python
print("DecisionTree Score:", dt.score(test_x_vector, test_y))
print("DecisionTree F1:", f1_score(test_y, dt.predict(test_x_vector), labels=['positive','negative'], average=None))
print(classification_report(test_y, dt.predict(test_x_vector), labels=['positive','negative']))
print("Confusion Matrix DT:")
print(confusion_matrix(test_y, dt.predict(test_x_vector), labels=['positive', 'negative']))
```

```python
# Comparación de modelos
models = ['SVM', 'LogisticRegression', 'GaussianNB', 'DecisionTree']
scores = [svc.score(test_x_vector, test_y), lr.score(test_x_vector, test_y), gnb.score(test_x_vector.toarray(), test_y), dt.score(test_x_vector, test_y)]
for model, score in zip(models, scores):
    print(f"{model}: {score:.4f}")
```
